```python
wine_list = {}
#,country,description,designation,points,price,province,region_1,region_2,variety,winery
with open("winemag10k.csv","r") as f_obj:
  reader = csv.DictReader(f_obj)
  for dictrow in reader:
    number = dictrow['']
    points = dictrow['points']
    price = dictrow['price']
    wine_list[number] = [points,price]
print(wine_list)
```

```python
coordinates_list = []
for k,v in wine_list.items():
    coordinates_list.append(v)
print(coordinates_list[0:10])
```

```python
x_price = []
y_points = []

for item in coordinates_list:
    x_price.append(item[1])
    y_points.append(item[0])
print(x_price[0:10])
print(y_points[0:10])
```
